```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('book (3).csv',encoding='latin-1')
df=df.iloc[:,1:]
df=df.sort_values(by='User.ID').reset_index(drop=True)
df
```

```python
# no. of unique values in each column
df.nunique()
```

```python
df[df.duplicated()]
```

```python
book=df.drop_duplicates().reset_index(drop=True)
```

```python
book[book.duplicated()]
```

```python
user_books=df.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index()
user_books
```

```python
user_books.index=book['User.ID'].unique()
user_books
```

```python
user_books.fillna(0,inplace=True)
user_books
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
```

```python
user_sim = 1 - pairwise_distances(user_books.values,metric='cosine')
user_sim
```

```python
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df.index = df['User.ID'].unique()
user_sim_df.columns = df['User.ID'].unique()
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[:20]
```
